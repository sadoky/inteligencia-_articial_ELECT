# Neuronal network 2D (very simple example)

#Librerias

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pl
import matplotlib as plt
import matplotlib.pyplot as plt
import random 
import operator
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
```

# Listas de entrada a la red 

```python
n=30                                         # numero total de elementos de las listas   

edad=[0]*n                                   # se crea el vector vacio con 30 elementos "edad"
for i in range (n):
  edad[i]=random.randint(1,25)               # se asignan valores al azar entre [1,25]


notas=[0]*n                                  # se crea el vector vacion con 30 elementos "notas"
for i in range (n):
  notas[i]=random.randint(1,10)              # se asignan valores al azar entre [1,10]
```

# Pesos y bias 

```python
item1=[0]*n
item2=[0]*n

for i in range(n):
  item1[i]=0.3          # peso #1
  item2[i]=-0.2         # peso #2


bias=[0]*n             # creacion para la lista llamada " bias "

for i in range(30):
  bias[i]=0.1
```

# Funcion de activacion 

```python
sum1=[0]*n
sum2=[0]*n
sum3=[0]*n

for i in range(n):
  sum1[i]=item1[i]*edad[i]
  sum2[i]=item2[i]*notas[i]
  sum3[i]=sum1[i]+sum2[i]+bias[i]     # Red neuronal en 2D
```

#decisión  (aprobado no aprobado)

```python
for i in range(n):  
  if sum3[i] > 0:           # condicional necesario para realizar la clasificación
    print("aceptado")
  else:
    print("no aceptado")
```

# VISUALIZACIÓN ENTORNO DE DECISIÓN 

```python

def f(x):
  return 0.5*x                         # función lineal 

x= list(range(1,31))                   # variable independoente de la función lineal 
y=[f(i) for i in (x)]                  # f(x)



# listas necesarias para realizar la clasifiacion en el gráfico 

aprobar2=[]
edad22=[]
edad4=[]
noa=[]




for i in range (n):
  for element in x:
    if element==edad[i]:
      if notas[i]>((element/2)):        # condicional comparativo para realizar la clasificacion por encima o debajo de la fución lineal 
        aprobar2.append(notas[i])
        edad22.append(edad[i])
      else:
        noa.append(notas[i])
        edad4.append(edad[i])





fig, ax = plt.subplots()
plt.plot(edad22,aprobar2,'o')
plt.plot(edad4,noa,'o')
ax.plot(y)
```

# NEURONAL NETWORK 3D 

```python


item3=[0]*n

for i in range(n):
  item3[i]=-0.1                          # peso para la tercer entrada



sum4=[0]*n
peso=[0]*n
for i in range (n):
  peso[i]=random.randint(1,40)           # variable de entrada de la red #3




for i in range(n):
  sum4[i]=item3[i]*peso[i]+sum3[i]     # red neuronal en 3D
```

# Frontera de Decision 

```python
for i in range(n):
  if sum4[i] > 0:
    print("aceptado")
  else:
    print("no aceptado")
```
