# XOR


```python
in1=[0,1]
in2=[1,0]


# y=-x+3    and
#y=-x+1.5   or 

peso1=1.7
peso2=1.8
bias1=-2    # bias intercepto 

peso3=3.9   # pesos pendiente de la recta 
peso4=3.5
bias2=-1

for i in range(2):
  for j in range(2):
    p1=(peso1*in1[i])+(peso2*in2[j])+bias1   # AND
    p2=(peso3*in1[i])+(peso4*in2[j])+bias2   # OR 
    

    if p1 > 0:
      AND=1
      NAND=0
    else:
      AND=0
      NAND=1

    if p2>0:
      OR=1
      
    else:
      OR=0
    

    p3=(OR*peso1)+(peso2*NAND)+bias1

    if p3>0:
      RES=1
      
    else:
      RES=0

    print(RES)
      





```
